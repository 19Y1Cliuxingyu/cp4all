**Object-Oriented Programming**
* class - general (templated - data and methods)
* It is a 'blueprint' created by a programmer for an object. It defines a set of attributes and methods that the object would have when initialized.
* object - specific (runtime instance) of a class
* encapsulation
  * group data and methods
  * information hiding - private data only accessible via public methods
* inheritance (is-a) code reuse
  * subclass adopt data and methods of superclass
  * subclass can also have its own data and methods, of which some are newly added or modified from the parent class
  * It is normally applied to classes
* polymorphism - code generality
  * ability to invoke different methods using the same name
  * dynamic binding
  * usually applied to functions and methods


```
import math

class Shape: # superclass / parent / base / ancestor

  # constructor method - function within a class
  def __init__(self, object_id):
    self.oid = object_id # private data (for some languages but not Python)

  # assessor - read/getty
  def get_oid(self): # information hiding
    return self.oid

  # modifier / mutator - change/setty
  def set_oid(self, new_oid):
    self.oid = new_oid

  # support / helper
  def show(self):
    print("OID:", self.oid)

  def calc_area(self):
    return "Dun know how to"


class Circle(Shape): # subclass

  def __init__(self, oid, radius):
    super().__init__(oid)
    self.radius = radius

  def get_radius(self):
    return self.radius

  def set_radius(self, new_radius):
    self.radius = new_radius

  def show(self):
    super().show()
    print("Radius:", self.radius)

  def calc_area(self):
    return math.pi * self.radius * self.radius


class Rectangle(Shape): # subclass

  def __init__(self, oid, length, breadth):
    super().__init__(oid)
    self.length = length
    self.breadth = breadth

  def get_length(self):
    return self.length

  def get_breadth(self):
    return self.breadth

  def set_length(self, new_length):
    self.length = new_length

  def set_breadth(self, new_breadth):
    self.breadth = new_breadth

  def show(self):
    super().show()
    print("Length:", self.length)
    print("Breadth:", self.breadth)    

  def calc_area(self):
    return self.length * self.breadth


# main
circle1 = Circle(3, 10)
rectangle1 = Rectangle(4, 8, 6)
#print(circle1.get_radius())
#print(rectangle1.get_length())
#print(rectangle1.get_breadth())
#circle1.set_radius(13)
#rectangle1.set_length(9)
#rectangle1.set_breadth(7)
#print(circle1.get_radius())
#print(rectangle1.get_length())
#print(rectangle1.get_breadth())
#circle1.show()
#rectangle1.show()

object1 = Shape(1)

shapes = [object1, circle1, rectangle1]
for shape in shapes:
  shape.show() # magic! polymorphism
  print(shape.calc_area())


#print(object1)
#print(object1.oid)
#print(object1.get_oid())
#object1.set_oid(999)
#print(object1.get_oid())
#object1.show()

#object2 = Shape(2)
#print(object2)
#print(object2.oid)
#print(object2.get_oid())
```

```
class Person:

  def __init__(self, person_id, person_name):
    self.pid = person_id
    self.name = person_name

  def get_pid(self):
    return self.pid

  def get_name(self):
    return self.person_name

  def set_name(self, new_name):
    self.name = new_name

  def display(self):
    print(self.pid)
    print(self.name)


class Student(Person):

  def __init__(self, person_id, person_name, classid):
    super().__init__(person_id, person_name)
    self.classid = classid

  def get_classid(self):
    return self.classid

  def set_classid(self, new_class):
    self.classid = new_class

  def display(self):
    super().display()
    print(self.classid)


class Staff(Person):

  def __init__(self, person_id, person_name, department):
    super().__init__(person_id, person_name)
    self.dept = department

  def get_dept(self):
    return self.dept

  def set_dept(self, new_department):
    self.dept = new_department

  def display(self):
    super().display()
    print(self.dept)


# main
person = Person(1, 'Tom')
#person.display()
student = Student(2, 'Mary', '5C35')
#student.display()
staff = Staff(3, 'Ali', 'Math')
#staff.display()

humans = [person, staff, student]
for human in humans:
  human.display()
```
